```python
X=dataset.iloc[:,:-1]
```

```python
X.head()
```

```python
type(X)
```

```python
y=dataset.iloc[:,3]
```

```python
y.head()
```

```python
type(y)
```

```python
Xtrain,X_test,y_train,y_test=train_test_split(X, y, random_state=1)
```

```python
lr=LinearRegression()
```

```python
lr.fit(Xtrain, y_train)
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
features=['TV','Radio','Newspaper']
```

```python
zip(features, lr.coef_)
```

```python
y_pred=lr.predict(X_test)
```

```python
#MAE - Mean Absolute Error - It is easy to understand, beacuase average error
mean_absolute_error(y_test,y_pred)
```

```python
#MSE - Mean Squared Error - It penalises for large errors
mean_squared_error(y_test, y_pred)
```

```python
#RMSE -Root Mean Squared Error- It is easily interpretable and its units are similar to units of response variable
np.sqrt(mean_squared_error(y_test,y_pred))
```

```python
round(lr.score(X_test, y_test)*100,2)
```
