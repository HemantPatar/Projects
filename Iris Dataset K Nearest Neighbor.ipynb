```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn=KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train, y_train)
```

```python
y_pred=knn.predict(X_test) 
```

```python
#Confusion Matrix
```

```python
confusion_matrix(y_test, y_pred)
```

```python
#Accuracy
```

```python
acc_score=accuracy_score(y_test, y_pred)*100
```

```python
str(round(acc_score, 2))
```

```python
#Using cross validation for parameter tuning
```

```python
from sklearn.model_selection import cross_val_score
```

```python
k_list=list(range(1, 50, 2))
```

```python
CV_Scores=[]
```

```python
#Using 10 fold Cross Validation 
```

```python
for k in k_list:
  knn = KNeighborsClassifier(n_neighbors=k)
  score=cross_val_score(knn, X_train, y_train, cv=10, scoring='accuracy')
  CV_Scores.append(score.mean())
```

```python
len(CV_Scores)
```

```python
#Finding Misclassification Error
```

```python
MCE=[1 - i for i in CV_Scores]
```

```python
len(MCE)
```

```python
#Plotting k values and their respective Misclassification errors
```

```python
plt.figure()
plt.figure(figsize=(15, 10))
plt.title("Optimal K Value", fontsize=25, fontweight='bold')
plt.xlabel("No. of Neighbors", fontsize=15)
plt.ylabel("Misclassification Error", fontsize=15)
sns.set_style('whitegrid')
plt.plot(k_list, MCE)

plt.show()
```

```python
#Finding best k
```

```python
best_k = k_list[MCE.index(min(MCE))]
```
