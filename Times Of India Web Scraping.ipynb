```python
import requests
from bs4 import BeautifulSoup
import re
from tqdm import tqdm
```

```python


urls = ['https://timesofindia.indiatimes.com/','https://timesofindia.indiatimes.com/briefs','https://timesofindia.indiatimes.com/briefs/india',
       'https://timesofindia.indiatimes.com/briefs/sports','https://timesofindia.indiatimes.com/briefs/entertainment','https://timesofindia.indiatimes.com/briefs/lifestyle',
       'https://timesofindia.indiatimes.com/briefs/gadgets','https://timesofindia.indiatimes.com/briefs/world','https://timesofindia.indiatimes.com/briefs/business']
```

```python
duplicate_urls=[]
main_urls=[]
article_urls=[]
for i in tqdm(urls):
    page=requests.get(i)
    contents=page.content
    soup=BeautifulSoup(contents, 'lxml')
    print(i)
    
    for link in soup.find_all('a'):
        duplicate_urls.append(link.get('href'))
```

```python
len(duplicate_urls)
```

```python
def Remove_duplicate(duplicate):
    for num in duplicate:
        if num not in main_urls:
            main_urls.append(num)
    return main_urls
```

```python
Remove_duplicate(duplicate_urls)
```

```python
def Remove_nonArticle(main_urls):
    for x in main_urls:
        if('/videos' not in x) and ('?' not in x) and ('/articlelist/' not in x) and('/photogallery.' not in x):
            if len(x)> 55:
                if x[0]=="/":
                    x="https://timesofindia.indiatimes.com" +x
                    print(x)
                    article_urls.append(x)
                else:
                    pass
            else:
                pass
        else:
            pass
```

```python
Remove_nonArticle(main_urls)
```

```python
all_data=[]

for i in tqdm(article_urls):
    Data={}
    try:
        page = requests.get(i)
        contents=page.content
        soup=BeautifulSoup(contents, 'lxml')
    except:
        pass
    try:
        Data['title']=soup.find('arttitle').text
    except:
        Data['title']= ""
        
    try:
        try:
            author = soup.find('div', class_='as_byline').text
            author=author.split("By -")
            author= author[1]
            author=author.split("Created")
            author=author[0]
            author=author.strip
            Data['author']=author
        except:
            author=soup.find('span', itemrop='author'.text)
            Data['author']=author
    except:
        Data['author']= ""
            
            
    try:
        date['article']=soup.find('div', class_='section1').text
        date = date.split("Updated")
        date=date[1]
        date=date.strip()
        Data['date'] = date
    except:
        Data['date']=""
    try:
        Data['article']=soup.find('div', class_='section1').text
    except:
        Data['article']= ""
    Data['website']="Times Of India"
    Data['media_type']="National"
    Data['language']="English"
    Data['url']=i
   
    all_data.append(Data)            
```
