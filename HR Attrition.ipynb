```python
dataset_category=dataset.select_dtypes(include=['object']).columns
```

```python
dataset_category
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
```

```python
dataset['Attrition']=le.fit_transform(dataset[dataset_category[0]])
```

```python
dataset['BusinessTravel']=le.fit_transform(dataset[dataset_category[1]])
dataset['Department']=le.fit_transform(dataset[dataset_category[2]])
dataset['EducationField']=le.fit_transform(dataset[dataset_category[3]])
dataset['Gender']=le.fit_transform(dataset[dataset_category[4]])
dataset['JobRole']=le.fit_transform(dataset[dataset_category[5]])
dataset['MaritalStatus']=le.fit_transform(dataset[dataset_category[6]])
dataset['Over18']=le.fit_transform(dataset[dataset_category[7]])
dataset['OverTime']=le.fit_transform(dataset[dataset_category[8]])
```

```python
dataset.dtypes
```

```python
dataset.columns
```

```python
dataset.head()
```

```python
pd.get_dummies(dataset, columns=['Attrition', 'BusinessTravel', 'Department', 'EducationField', 'Gender',
       'JobRole', 'MaritalStatus', 'Over18', 'OverTime'], drop_first=True)
```

```python
temp=pd.get_dummies(dataset, columns=['Attrition', 'BusinessTravel', 'Department', 'EducationField', 'Gender',
       'JobRole', 'MaritalStatus', 'Over18', 'OverTime'], drop_first=True)
```

```python
dataset.drop(['Attrition', 'BusinessTravel', 'Department', 'EducationField', 'Gender',
       'JobRole', 'MaritalStatus', 'Over18', 'OverTime'], axis=1, inplace= True)
```

```python
dataset.shape
```

```python
dataset_final=pd.concat([dataset, temp], axis=1)
```

```python
dataset_final.shape
```

```python
dataset_final.columns
```

```python
X=dataset_final.drop(['Attrition_1'], axis=1)
```

```python
y=dataset_final['Attrition_1']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test=train_test_split(X, y, test_size=0.8)
```

```python
from sklearn.ensemble import RandomForestClassifier

rf=RandomForestClassifier()
rf.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score
test_pred=rf.predict(X_test)
```

```python
accuracy_score(y_test, test_pred)
```

```python

feat_importances = pd.Series(rf.feature_importances_, index=X.columns)
feat_importances = feat_importances.nlargest(20)
feat_importances.plot(kind='barh')
```
